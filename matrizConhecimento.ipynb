Implementing Snowball stem...

```python
from nltk.stem.snowball import SnowballStemmer
snow_stemmer = SnowballStemmer(language='portuguese')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# * Tokenizing and removing punctuation
data = materias_unb.loc[3]['Emenda/Descrição']
# data = 'Flor, floração, flora, floral, floreado, floreio, flóreo, florescente, floriano, floricultura, florido, florista, florzinha'
test_words = word_tokenize(data, language='portuguese')
test_words=[word.lower() for word in test_words if word.isalpha()]

# * removing stop words
non_stop_words = [word for word in test_words if not word in stopwords.words('portuguese')]
print(non_stop_words)

# * Stemming
stem_words = []
for word in non_stop_words:
    x = snow_stemmer.stem(word)
    stem_words.append(x)

# * Output
for w1, w2 in zip(non_stop_words, stem_words):
    print(w1 + ' -> ' + w2)
```
